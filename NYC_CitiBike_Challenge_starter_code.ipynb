```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
file_1 = "2013-08 - Citi Bike trip data.csv"
file_2 = "2014-02 - Citi Bike trip data.csv"
df1 = pd.read_csv(file_1)
df2 = pd.read_csv(file_2)
```

```python
# Merging dataframes
df_merged = pd.concat([df1, df2], ignore_index = True)
df_merged.head()
print(df_merged.columns)
```

```python
# assuming that the tripduration column is in seconds
df_merged['tripminutes'] = df_merged['tripduration'] / 60
df_merged['tripminutes'] = df_merged['tripminutes'].round()

# move the tripminutes column to the beginning of the dataframe
cols = df_merged.columns.tolist()
cols = ['tripminutes'] + cols[:-1]
df_merged = df_merged[cols]
df_merged.tail()
```

```python
# 2. Check the datatypes of your columns. 
df_merged.info()
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.

df_merged["tripduration"] = pd.to_datetime(df_merged["tripduration"])
```

```python
# 4. Check the datatypes of your columns. 
df_merged.info()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
df_merged.to_csv("merged.csv", index = False)
```
